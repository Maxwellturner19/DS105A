## Part 1:

Install Libraries required 

```python
import pandas as pd
import requests 
from pprint import pprint
from scrapy import Selector
```

Request URL:

```python

my_url = 'https://socialdatascience.network/#schedule'


response = requests.get(my_url)

response
```

Begin to parse HTML

```python
sel = Selector(text=response.text)
```

```python
cards = sel.css("div.card-body")
```

```python
title = cards.css('a h6 ::text').getall()
```

```python
dates_speakers = cards.css('div.card-body p ::text').getall()  # dates and speakers found together, so I split them 
dates = dates_speakers[1::2]
speakers = dates_speakers[::2]
```

```python
link = cards.css('a ::attr(href)').getall()
link = link[::2]
```

```python
link_suffix = cards.css("a ::attr(href)").getall()
link_suffix

link_agenda= []
for i in range(len(link_suffix)):
    suffix =link_suffix[i]
    link_agenda.append('https://socialdatascience.network/' + suffix)

print(len(link_agenda))          # this is wrong 
```

Count is wrong 

```python
link_agenda = link_agenda[::2]  # I index to make correct length after observation
len(title), len(dates), len(speakers), len(link_agenda)
```

count is all good. Need to tidy up the DataFrame

```python
for i in range (len(dates)):  # I remove the "Date: Wednesday"
    dates[i] = dates[i].replace("Date: Wednesday, ", "")

for i in range(len(speakers)):  # I remove the "Speaker:" 
    speakers[i] = speakers[i].replace("Speaker: ", "")

df_schedule = pd.DataFrame({'Title': title, 'Speaker': speakers, 'Date': dates, 'Link': link_agenda})
df_schedule. to_csv('schedule.csv', index=False )
df_schedule
```

## Part 2:

Using part 1:

```python
for i in range(len(link_agenda)):
    response = requests.get(link_agenda[i])
    sel = Selector(text=response.text)
```

Parsing HTML again:

```python
specific_row = sel.css(".row.schedule-item")
```

```python
specific_title = specific_row.css('h4 ::text').getall()
```

```python
specific_description = specific_row.css('p ::text').getall()
```

```python
specific_speaker = specific_row.css('span ::text').getall()
```

```python
all_titles = []
all_descriptions = []
all_links = []  
all_speakers = []
```

create for loop to fill "all_variables"

create for loop to make length of "all_variables" same length and not distort the positioning of elements 

```python
for link in link_agenda:
    response = requests.get(link)
    sel = Selector(text=response.text)
    
    specific_row = sel.css(".row.schedule-item")
    
    specific_title = specific_row.css('h4::text').getall()
    specific_description = specific_row.css('p::text').getall()
    specific_speaker = specific_row.css('span::text').getall()

    all_links.extend([link] * len(specific_title))
    
    all_titles.extend(specific_title) 
    all_descriptions.extend(specific_description)
    all_speakers.extend(specific_speaker)
    
    if len(specific_speaker) < len(specific_title):  # I add a for loop such that if the specific_speaker is shorter than the specific_title, I add empty strings to the specific_speaker list
        all_speakers.extend([""] * (len(specific_title) - len(specific_speaker))) # This means the array will be the same length so a dataframe can be made
```

Create DataFrame for 'schedule.csv'

```python
df_schedule = pd.DataFrame({'event_link': all_links, 'talk_title': all_titles, 'talk_speakers': all_speakers, 'event_description': all_descriptions}) 
df_schedule.to_csv('agenda.csv', index=False )
df_schedule
```
